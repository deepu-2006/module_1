```python
number_1 = int(input("enter the first number: "))
# Takes the first number from the user and converts it to an integer

sign_1 = input("enter sign: ")
# Takes the mathematical operator (+, -, *, /, **) from the user

number_2 = int(input("enter the second number: "))
# Takes the second number from the user and converts it to an integer

if sign_1 == "+":
    # Checks if the operator is addition
    print("addition is:", number_1 + number_2)
    # Adds both numbers and prints the result

elif sign_1 == "-":
    # Checks if the operator is subtraction
    print("subtract is:", number_1 - number_2)
    # Subtracts second number from first and prints the result

elif sign_1 == "*":
    # Checks if the operator is multiplication
    print("multiplication is:", number_1 * number_2)
    # Multiplies both numbers and prints the result

elif sign_1 == "/":
    # Checks if the operator is division
    print("division is:", number_1 / number_2)
    # Divides first number by second and prints the result

elif sign_1 == "**":
    # Checks if the operator is power
    print("square of the number:", number_1 ** number_2)
    # Raises first number to the power of second number and prints the result
```
